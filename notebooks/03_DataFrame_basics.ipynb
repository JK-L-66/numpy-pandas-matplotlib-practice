# DataFrame Basics

This notebook covers the fundamental concepts of Pandas DataFrames, including creation, indexing, attributes, statistical analysis, and practical operations.

---

## 1. Creating DataFrames

```python
import pandas as pd
import numpy as np

# --- Creating DataFrame from Series ---
s1 = pd.Series([1, 2, 3, 4, 5])
s2 = pd.Series([6, 7, 8, 9, 10])
df = pd.DataFrame({1: s1, 2: s2})
print("DataFrame from 2 Series:")
display(df)
print("Column type:", type(df[1]))

# --- Creating DataFrame from a Dictionary ---
df = pd.DataFrame({
    "name": ["a", "a", "b", "c", "d", "e"],
    "age": [15, 15, 17, 20, 26, 30],
    "score": [60.5, 60.5, 80, 30.6, 70, 83.5]
}, index=[1, 2, 3, 4, 5, 6], columns=["name", "score", "age"])

print("\n" + "="*40)
print("Final DataFrame:")
display(df)
```

## 2. DataFrame Attributes

```python
print("Index:", df.index)
print("Columns:", df.columns)
print("Values:\n", df.values)
print("-" * 30)

print("ndim (dimensions):", df.ndim)
print("shape (rows, cols):", df.shape)
print("size (total elements):", df.size)
print("dtypes:\n", df.dtypes)
```

## 3. Indexing: Rows (`loc` vs `iloc`)

```python
# Accessing rows
print("df.loc[4] (Row with label 4):")
print(df.loc[4])
print("\ndf.iloc[3] (Row at integer position 3):")
print(df.iloc[3]) # Outputs the same as row 4 in this example
```

## 4. Column Access and Single Value Retrieval

```python
# Accessing a single column (returns a Series)
print("df.loc[:, 'name']:")
print(df.loc[:, "name"])
print("\ndf.iloc[:, 0] (First column by position):")
print(df.iloc[:, 0])

print("-" * 30)

# Getting single scalar values
print("df.at[3, 'score']:", df.at[3, 'score'])  # Access by label
print("df.iat[2, 1]:", df.iat[2, 1])            # Access by integer position (Row 2, Col 1)
print("df.loc[3, 'score']:", df.loc[3, 'score'])
print("df.iloc[2, 1]:", df.iloc[2, 1])
```

### 4.1 Accessing Columns: `df['col']` vs `df[['col']]`

```python
# Single bracket returns a Series
print("df['name'] (Series):")
print(df['name'])
print("Type:", type(df['name']))

print("\ndf.name (Series - attribute style):")
print(df.name)

print("\ndf[['name']] (DataFrame with single column):")
print(df[['name']])
```

## 5. Exploring Data: Head, Tail, and Sampling

```python
print("df.head(2):")
print(df.head(2))
print("\ndf.tail(3):")
print(df.tail(3))
print("\ndf.sample(3) (Randomly select 3 rows):")
print(df.sample(3))
```

## 6. Boolean Masking and Conditional Filtering

```python
# Boolean mask: Score > 70
print("Boolean mask (df.score > 70):")
print(df.score > 70)

# Compound mask: Score > 70 and Age < 30
print("\nFiltered rows (Score > 70 & Age < 30):")
print(df[(df["score"] > 70) & (df.age < 30)])

print("\nCheck if values are in list ['b', 20]:")
print(df.isin(['b', 20]))
print("\nCheck for NaN values:")
print(df.isna())
```

## 7. Descriptive Statistics and Aggregations

```python
print("Sum of scores:", df['score'].sum())
print("Max score:", df['score'].max())
print("Min age:", df.age.min())
print("Mean score:", df['score'].mean())
print("Median score:", df['score'].median())
print("Mode of age:", df['age'].mode())
print("-" * 30)

print("Std dev of score:", df.score.std())
print("Variance of score:", df.score.var())
print("25th Percentile of score:", df.score.quantile(0.25))
print("-" * 30)

print("Full descriptive statistics:")
print(df.describe())
print("\nCount of non-null values per column:")
print(df.count())

print("\nValue counts (all columns together):")
print(df.value_counts())
```

## 8. Deduplication and Data Transformation

```python
print("DataFrame after dropping exact duplicate rows:")
print(df.drop_duplicates())

print("\nDuplicates based on 'age' column:")
print(df.duplicated(subset=['age']))

print("-" * 30)

# Replacing values
print("Replace age 15 with 39:")
print(df.replace(15, 39))

# Cumulative operations
print("\nCumulative max (score):")
display(df.cummax(axis=0))
```

## 9. Sorting and Selecting Top Rows

```python
print("Sort by index (Descending):")
print(df.sort_index(ascending=False))

print("\nSort by multiple columns (Score Asc, Age Desc):")
print(df.sort_values(by=['score', 'age'], ascending=[True, False]))

print("\nTop 2 scores overall (nlargest):")
print(df.nlargest(2, columns=['score', 'age']))

print("\nBottom 2 scores overall (nsmallest):")
print(df.nsmallest(2, columns=['score', 'age']))
```

## 10. Practical Case Study: Student Scores Analysis

```python
data = {
    'name': ['a', 'b', 'c', 'd', 'e'],
    'math': [85, 92, 78, 88, 95],
    'english': [90, 88, 85, 92, 80],
    'physics': [75, 80, 88, 85, 90]
}
scores = pd.DataFrame(data)

print("Raw student scores:")
display(scores)

# Adding new columns
scores['sum'] = scores[['math', 'english', 'physics']].sum(axis=1)
scores['average'] = scores[['math', 'english', 'physics']].mean(axis=1)

print("\nDataFrame with 'sum' and 'average' columns:")
display(scores)

print("\nFilter: Math > 90 OR English > 60:")
display(scores[(scores['math'] > 90) | (scores['english'] > 60)])

print("\nTop 3 students by total score:")
print(scores.sort_values('sum', ascending=False).head(3))

# Alternative way to find top 3
print("\nTop 3 students using nlargest:")
print(scores.nlargest(3, columns=['sum']))
```

## 11. Practical Case Study: Product Sales

```python
data = {
    'product': ['A', 'B', 'C', 'D'],
    'price': [100, 150, 200, 120],
    'num': [50, 30, 20, 40]
}
df = pd.DataFrame(data)

# Calculate total revenue for each product
df['total'] = df['price'] * df['num']

print("Product sales list with total revenue:")
display(df)

print("\nBest selling product by total revenue:")
print(df.nlargest(1, columns=['total']))

print("\nFull list sorted by total revenue (Descending):")
print(df.sort_values('total', ascending=False))
```
