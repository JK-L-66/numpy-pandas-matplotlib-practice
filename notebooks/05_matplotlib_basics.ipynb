# Matplotlib Basics

## 1. Line Chart with Annotations

```python
import matplotlib.pyplot as plt

# Data preparation
months = ['Jan', 'Feb', 'Mar', 'Apr']
sales = [100, 150, 80, 130]

# Create figure
plt.figure(figsize=(10.0, 5.0))

# Plot line with marker and label
plt.plot(months, sales, label='Product A', linewidth=2, marker='o')

# Styling and labels
plt.title('Monthly Sales Performance', color='blue', fontsize=20)
plt.xlabel('Month', color='blue', fontsize=12)
plt.ylabel('Sales (in USD)', color='blue', fontsize=12)
plt.legend(loc='upper left', fontsize=12)
plt.grid(True, alpha=0.3, color='gray', linestyle='--')
plt.ylim([0, 160])

# Annotate data points directly on the graph
for x, y in zip(months, sales):
    plt.text(x, y, str(y), ha='center', va='bottom', fontsize=10)

plt.show()
```

## 2. Pie Chart

```python
# Data
subjects = ['Chinese', 'Math', 'English', 'Biology']
scores = [99, 44, 88, 100]

# Create pie chart
plt.figure(figsize=(6, 6))
# autopct adds percentage labels automatically
plt.pie(scores, labels=subjects, autopct='%1.1f%%')

plt.title('Score Distribution', fontsize=16)
plt.show()
```

## 4. Feature Engineering and Plotting

```python
# Create a new feature: daily mean temperature
df_2015['temp_mean'] = (df_2015['temp_max'] + df_2015['temp_min']) / 2
print("New column 'temp_mean' created:")
df_2015[['date', 'temp_mean']].head()

# Plot the mean temperature
plt.figure(figsize=(15.0, 6.0))
plt.plot(df_2015['date'], df_2015['temp_mean'], label='Mean Temperature', color='orange')

plt.title('2015 Daily Mean Temperature', fontsize=18)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Temperature (°C)', fontsize=12)
plt.legend()

plt.show()
```

## 5. Histogram (Exploring Data Distribution)

```python
# Visualize the distribution of precipitation
plt.figure(figsize=(8, 5))

plt.hist(df_2015['precipitation'], bins=5, color='steelblue', edgecolor='black')

plt.title('Distribution of Daily Precipitation (2015)', fontsize=16)
plt.xlabel('Precipitation Amount', fontsize=12)
plt.ylabel('Frequency (Number of Days)', fontsize=12)

plt.show()
```
