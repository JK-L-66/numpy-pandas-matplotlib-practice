# Pandas Basics

## 1. Series: Creation and Basic Attributes

```python
import pandas as pd
import numpy as np

# Creating a Series with custom index and name
s = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'], name='month')
s1 = pd.Series(s, index=['a', 'c'])

print("Original Series 's':")
print(s)
print("\nFiltered Series 's1' (indices 'a', 'c'):")
print(s1)
```

## 2. Series Attributes and Indexing (`loc` vs `iloc`)

```python
print("Index:", s.index)
print("Values:", s.values)
print("Shape, ndim, size:", s.shape, s.ndim, s.size)
print("Data type (dtype):", s.dtype)

# Accessing elements by label (loc) and integer position (iloc)
print("s.loc['b'] (Label):", s.loc['b'])
print("s.iloc[1] (Position):", s.iloc[1])

# Slicing
print("s.loc['a':'c']:")
print(s.loc['a':'c'])
print("s.iloc[1:2]:")
print(s.iloc[1:2])

# Fast scalar access
print("s.at['a']:", s.at['a'])
print("s.iat[0]:", s.iat[0])

# Boolean masking
print("s[ s < 3 ]:")
print(s[s < 3])

# Head and Tail
print("s.head():")
print(s.head())
print("s.tail():")
print(s.tail())
```

## 3. Handling Missing Values (`NaN`) and Descriptive Statistics

```python
# Creating a Series with `NaN` values
s = pd.Series([10, 2, np.nan, None, 3, 4, 5], index=['A', 'B', 'C', 'D', 'E', 'F', 'G'], name='data')
print("Series with NaN:")
print(s)

print("\n--- Basic Info ---")
print("Non-NaN count:", s.count())
print("Index keys:", s.keys())
print("Index:", s.index)
print("Check for NaNs:")
print(s.isna())
print("Check if values are in [4, 5, 6]:")
print(s.isin([4, 5, 6]))

print("\n--- Statistical Methods (NaN excluded automatically) ---")
print("Mean:", s.mean())
print("Sum:", s.sum())
print("Std deviation:", s.std())
print("Variance:", s.var())
print("Min:", s.min())
print("Max:", s.max())
print("Median:", s.median())
print("80th Percentile:", s.quantile(0.8))
# Important fix: Added parentheses `()` to call the method
print("Mode:", s.mode())

print("\n--- Frequency & Value Counts ---")
print(s.value_counts())
```

## 4. Sorting, Unique, and Deduplication

```python
print("Sort by values:")
print(s.sort_values())
print("\nSort by index:")
print(s.sort_index())

# Unique values
print("Unique values (method):")
print(s.unique())
print("Number of unique values:", s.nunique())

# Drop duplicates
print("\nDrop duplicates:")
print(s.drop_duplicates())
```

## 5. Practical Case Studies - Students & Temperature

```python
# Case 1: Students Scores
np.random.seed(42)
values = np.random.randint(50, 101, 10)
indexes = ['student' + str(i) for i in range(1, 11)]
scores = pd.Series(values, index=indexes)

print("Students scores:")
print(scores)
print("\nMean:", scores.mean())
print("Max:", scores.max())
print("Min:", scores.min())

above_avg = scores[scores > scores.mean()]
print("Count above average:", len(above_avg))
print("Scores above average:")
print(above_avg)

print("-" * 40)

# Case 2: Weekly Temperatures
temperatures = pd.Series([28, 31, 29, 32, 30, 27, 33],
                          index=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])

print("Temperatures > 30 (Days count):", temperatures[temperatures > 30].count())
print("Mean temperature:", temperatures.mean())

print("\nTemperatures sorted (Descending):")
print(temperatures.sort_values(ascending=False))

# Identify days with highest absolute temperature change compared to previous day
diff_abs = temperatures.diff().abs()
print("\nLargest temperature changes (sorted):")
print(diff_abs.sort_values(ascending=False))
```

```python
# Calculate absolute temperature difference between consecutive days
t3 = temperatures.diff().abs()
# Get the two days with the largest temperature differences
highest_diff_days = t3.sort_values(ascending=False).keys()[:2].tolist()
print("Days with largest temperature difference:", highest_diff_days)
```

## 6. Time Series Analysis - Stocks & Sales

```python
# Stock prices example
prices = pd.Series([102.3, 103.5, 105.1, 104.8, 106.2, 107.0, 106.5, 108.1, 109.3, 110.2],
                    index=pd.date_range('2025-01-01', periods=10))
print("Stock Prices:")
print(prices)

# Percentage change (returns) calculation
a = prices.pct_change()

# Important fix: Added `()` to call the bound methods
print("Date of max growth:", a.idxmax())
print("Date of max drop:", a.idxmin())
print("Std dev of returns:", a.std())
```

```python
# Monthly sales example
sales = pd.Series([120, 135, 145, 160, 155, 170, 180, 175, 190, 200, 210, 220],
                   index=pd.date_range('2022-01-01', periods=12, freq='MS'))
print("Monthly Sales:")
print(sales)

print("\nMonth with max sales:", sales.idxmax())

# Find 3 consecutive months with sales growth
change = sales.pct_change()
positive_growth = change > 0
# Rolling window of 3, checking if sum is 3 (meaning all are True)
consecutive_growth_dates = positive_growth[positive_growth.rolling(3).sum() == 3].keys().tolist()
print("Starting months of 3 consecutive sales growth:", consecutive_growth_dates)
```

## 7. Data Resampling and Advanced Filtering

```python
# Quarterly resampling (average)
print("Quarterly sales average:")
print(sales.resample('QS').mean())

print("\nQuarterly sales count:")
print(sales.resample('QS').count())

print("-" * 40)

# Hourly sales simulation
np.random.seed(42)
hours_sales = pd.Series(np.random.randint(0, 100, 24),
                        index=pd.date_range('2026-01-01', periods=24, freq='h'))

# Extract business hours (8:00 - 22:00)
is_business = (hours_sales.index.hour >= 8) & (hours_sales.index.hour <= 22)
business_hours_sales = hours_sales[is_business]
non_business_sales = hours_sales[~is_business]

total_business = business_hours_sales.sum()
total_non_business = non_business_sales.sum()

print(f"Total sales during business hours: {total_business}")
print(f"Total sales during non-business hours: {total_non_business}")
print(f"Ratio (Business / Non-Business): {total_business / total_non_business:.4f}")

print("\nTop 3 hourly sales (Date/Time indices):")
print(hours_sales.nlargest(3).keys())
```
