## 2. Missing Values Detection and Imputation

```python
# --- Test different kinds of missing values ---
s = pd.Series([1, 2, np.nan, None, pd.NA])
print("Series with mixed missing values:")
print(s)
print("\nCheck NaN (isna()):")
print(s.isna())
print("Sum of NaNs:", s.isna().sum())

print("-" * 40)

# --- Load weather dataset with NaNs ---
weather_withna = pd.read_csv('../data/weather_withna.csv', sep=',')
print("Weather dataset preview:")
display(weather_withna)

print("\nSum of NaNs per column:")
print(weather_withna.isna().sum())

# --- Methods to handle missing values ---
print("\nDrop rows where ALL values are NaN (how='all'):")
print(weather_withna.dropna(how='all').tail())

print("\n--- Fill with fixed values (temp_max=20, wind=2.5) ---")
print(weather_withna.fillna({'temp_max': 20, 'wind': 2.5}).tail())

print("\n--- Fill with the column mean ---")
print(weather_withna.fillna(weather_withna[['temp_max', 'wind']].mean()).tail())

print("\n--- Forward Fill (ffill) - Use previous row's valid value ---")
print(weather_withna.ffill().tail())

print("\n--- Backward Fill (bfill) - Use next row's valid value ---")
print(weather_withna.bfill().tail())
```

## 3. Deduplication, Data Types and Binning

```python
# --- Duplicates ---
data = {
    "name": ['alice', 'alice', 'bob', 'alice', 'jack', 'bob'],
    "age": [26, 25, 30, 25, 35, 30],
    'city': ['NY', 'NY', 'LA', 'NY', 'SF', 'LA']
}
df = pd.DataFrame(data)
print("Original dataframe:")
display(df)

print("\nCheck duplicates:")
print(df.duplicated())
print("\nDrop duplicates based on 'name' (keeping first):")
print(df.drop_duplicates(subset=['name']))

print("-" * 40)

# --- Type Conversion & Mapping ---
df = pd.read_csv('../data/sleep.csv')

# Convert dtype to category
df['gender'] = df['gender'].astype('category')
print("Gender column dtype after conversion:")
print(df['gender'].dtype)

# Create a boolean category mapping
df['is_male'] = df['gender'].map({'Female': True, 'Male': False})
print("\nMapped column 'is_male':")
print(df['is_male'].head())

# --- Discretization (Binning) with salary data ---
df_emp = pd.read_csv('../data/employees.csv')
df1 = df_emp.head(10)[['employee_id', 'salary']]

# Cut into 3 equal-width bins
print("\nSalary bins (equal width):")
print(pd.cut(df1['salary'], bins=3).value_counts())

# Cut with custom bins labels
df1['salary_level'] = pd.cut(df1['salary'], bins=[0, 10000, 20000, 30000],
                               labels=['Low', 'Medium', 'High'])
print("\nSalary level categories (Custom bins):")
print(df1['salary_level'])
```

## 4. Reshaping Data: Melt, Pivot, and String Splitting

```python
# --- Wide to Long (Melt) ---
data = {
    'ID': [1, 2],
    'name': ['alice', 'bob'],
    'Math': [90, 85],
    'English': [88, 92],
    'Science': [95, 89]
}
df = pd.DataFrame(data)
print("Original Wide Format:")
display(df)

# Melt
df2 = pd.melt(df, id_vars=['ID', 'name'], var_name='subject', value_name='note')
df2 = df2.sort_values('name')
print("\nMelted to Long Format:")
print(df2)

# Pivot back to Wide
print("\nPivot back to Wide Format:")
display(pd.pivot(df2, index='ID', columns='subject', values='note'))

print("-" * 40)

# --- String Splitting ---
df_sleep = pd.read_csv('../data/sleep.csv')
df_bp = df_sleep[['person_id', 'blood_pressure']].copy()

# Split blood_pressure '120/80' into '120' and '80'
df_bp[['high', 'low']] = df_bp['blood_pressure'].str.split('/', expand=True)
df_bp['high'] = df_bp['high'].astype('int64')
df_bp['low'] = df_bp['low'].astype('int64')

print("Blood Pressure split into high/low integers:")
print(df_bp.head())
```

## 5. Time Series Processing (Timestamps and Resampling)

```python
# --- Timestamp properties ---
d = pd.Timestamp('2015-05-02 10:22')
print("Timestamp:", d)
print("Type:", type(d))
print(f"Year: {d.year}, Quarter: {d.quarter}, Month: {d.month}, Day: {d.day}")
print(f"Hour: {d.hour}, Minute: {d.minute}, Second: {d.second}")
print("Is end of month:", d.is_month_end)
print("Day of the week (Monday=0):", d.day_of_week)
print("Day name:", d.day_name())

print("-" * 40)

# --- Resampling Weather Data ---
df_weather = pd.read_csv('../data/weather.csv', parse_dates=['date'])
df_weather.set_index('date', inplace=True)

print("Monthly average temperatures (using 'ME' - Month End):")
display(df_weather[["temp_max", "temp_min"]].resample("ME").mean())

print("Monthly average temperatures (using 'MS' - Month Start):")
display(df_weather[["temp_max", "temp_min"]].resample("MS").mean())

# --- Generating Date Ranges ---
days = pd.date_range("2025-07-03", periods=10, freq="W")
print("\nGenerated weekly date range:")
print(days)
```

## 6. GroupBy and Aggregation

```python
df = pd.read_csv('../data/employees.csv')
# Drop rows with missing department_id and convert to int
df = df.dropna(subset=['department_id'])
df['department_id'] = df['department_id'].astype('int64')

# 1. Groups by department_id
print("Grouped departments (keys):")
print(df.groupby('department_id').groups.keys())

print("\n--- Average salary per Department ---")
df2 = df.groupby('department_id')[['salary']].mean()
df2['salary'] = df2['salary'].round(2)
display(df2)

print("\nAverage salary by Department (Sorted descending):")
print(df2.sort_values('salary', ascending=False))

print("\n--- Multi-level GroupBy (Department + Job ID) ---")
df3 = df.groupby(['department_id', 'job_id'])[['salary']].mean()
df3['salary'] = df3['salary'].round(1)
print(df3.sort_values('salary', ascending=False))
```

## 7. Comprehensive Case Study: Penguins & Sleep Analysis

```python
# --- Penguins Case ---
df_penguins = pd.read_csv('../data/penguins.csv')
print("Penguins dataset info:")
df_penguins.info()

print("\nCheck missing values:")
print(df_penguins.isna().sum())

# Drop rows with NaN
df_penguins.dropna(inplace=True)

# Create a new feature: bill ratio
df_penguins['bill_ratio'] = df_penguins['bill_length_mm'] / df_penguins['bill_depth_mm']

# Bin body mass into categories
labels = ['low', 'medium', 'high']
df_penguins['mass_level'] = pd.cut(df_penguins['body_mass_g'], bins=3, labels=labels)

print("\nBody mass distribution:")
print(df_penguins['mass_level'].value_counts())

# Aggregation by sex and island
print("\nMean body mass grouped by Sex & Island:")
display(df_penguins.groupby(['sex', 'island']).agg({
    'body_mass_g': ['mean', 'count']
}))

print("-" * 60)

# --- Sleep Data Case ---
df_sleep = pd.read_csv('../data/sleep.csv')

# Drop sleep_disorder column
df_sleep.drop(columns='sleep_disorder', inplace=True)

# Convert Categorical columns
df_sleep['gender'] = df_sleep['gender'].astype('category')
df_sleep['occupation'] = df_sleep['occupation'].astype('category')
df_sleep['bmi_category'] = df_sleep['bmi_category'].astype('category')

# Binning Age
age_labels = ['Youth', 'Middle', 'Old']
df_sleep['age_level'] = pd.cut(df_sleep['age'], bins=3, labels=age_labels)

print("BMI Category counts:")
print(df_sleep['bmi_category'].value_counts())

print("\nAggregated Sleep Statistics by Age Group & BMI Category:")
display(df_sleep.groupby(['age_level', 'bmi_category']).agg({
    'sleep_duration': 'mean',
    'sleep_quality': 'mean',
    'stress_level': 'mean'
}))
```
